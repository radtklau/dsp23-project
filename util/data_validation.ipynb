```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

import tensorflow as tf
import tensorflow_data_validation as tfdv
import pandas as pd
import json

data = '../data/folder_C/file_0.csv'
train_data = '../data/train.csv'
```

```python
def compute_visualize_stats(data: str, view:False):
    stats = tfdv.generate_statistics_from_csv(data_location=data)
    if view:
        tfdv.visualize_statistics(stats)
    return stats
```

```python
def infer_schema_from_train(display=False):
    train_stats = tfdv.generate_statistics_from_csv(data_location=train_data)
    schema = tfdv.infer_schema(train_stats)
    tfdv.display_schema(schema)
    return schema
```

### Visualizing stats for data sample

```python
data_stats = compute_visualize_stats(data, True)
```

### Comparing Schema to Data

```python
schema = infer_schema_from_train()
anomalies = tfdv.validate_statistics(statistics=data_stats, schema=schema)
tfdv.display_anomalies(anomalies)
```
